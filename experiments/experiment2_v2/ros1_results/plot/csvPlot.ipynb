```python
import  pandas as pd
import  matplotlib.pyplot as plt
import  csv
```

```python
sample_data = pd.read_csv('mean_times.csv')
```

```python
plt.plot(sample_data.id,sample_data.mean_time_ms_1)
plt.plot(sample_data.id,sample_data.mean_time_ms_10)
plt.plot(sample_data.id,sample_data.mean_time_ms_100)
plt.plot(sample_data.id,sample_data.mean_time_ms_200)
plt.plot(sample_data.id,sample_data.mean_time_ms_1000)
plt.plot(sample_data.id,sample_data.mean_time_ms_2000)
plt.plot(sample_data.id,sample_data.mean_time_ms_5000)
plt.plot(sample_data.id,sample_data.mean_time_ms_10000)
plt.plot(sample_data.id,sample_data.mean_time_ms_15000)
plt.plot(sample_data.id,sample_data.mean_time_ms_20000)

plt.legend(["1","10","100","200","1000","2000","5000","10000","15000","20000"])

plt.rcParams['savefig.dpi'] = 300 #图片像素
plt.rcParams['figure.dpi'] = 300 #分辨率

plt.title("Message Latency")
plt.xlabel("Message id")
plt.ylabel("Message latency (ms)")
plt.show()
# plt.savefig('plottest.png', dpi=300) #指定分辨率保存
```
