```python
"""csv"""
def write_raw_index(file):
    filename = file
    with open(filename, 'r+', encoding='utf-8') as f:
        content = f.read()
        f.seek(0, 0)
        #mid, text, source, uid
        text = "id" + "," + "message_latency"
        f.write(text + '\n' + content)
```

```python
def changeFileToRead(filename,filechangename):
    with open(filename,"r+") as f:
        f_csv=csv.reader(f)
        with open(filechangename,"w+",newline='') as f_1:
            f_1_csv=csv.writer(f_1)
            f_1_csv.writerow(["id","message_latency"])
            for row in f_csv:
                intMessage = int(row[1])/1000000
                # print(intMessage)
                f_1_csv.writerow([row[0],str(intMessage)])

            f_1.close()
        f.close()
```

```python
filename = 'times_2000.csv'
filechangename = '2000.csv'
changeFileToRead(filename,filechangename)
filename = 'times_1000.csv'
filechangename = '1000.csv'
changeFileToRead(filename,filechangename)
filename = 'times_10.csv'
filechangename = '10.csv'
changeFileToRead(filename,filechangename)
filename = 'times_1.csv'
filechangename = '1.csv'
changeFileToRead(filename,filechangename)
filename = 'times_100.csv'
filechangename = '100.csv'
changeFileToRead(filename,filechangename)
filename = 'times_200.csv'
filechangename = '200.csv'
changeFileToRead(filename,filechangename)
```

```python
sample_data1 = pd.read_csv('1.csv')
sample_data10 = pd.read_csv('10.csv')
sample_data100 = pd.read_csv('100.csv')
sample_data200 = pd.read_csv('200.csv')
sample_data1000 = pd.read_csv('1000.csv')
sample_data2000 = pd.read_csv('2000.csv')
plt.plot(sample_data1.id,sample_data1.message_latency)
plt.plot(sample_data10.id,sample_data10.message_latency)
plt.plot(sample_data100.id,sample_data100.message_latency)
plt.plot(sample_data200.id,sample_data200.message_latency)
plt.plot(sample_data1000.id,sample_data1000.message_latency)
plt.plot(sample_data2000.id,sample_data2000.message_latency)
plt.legend(["1","10","100","200","1000","2000"])


plt.rcParams['savefig.dpi'] = 300 #
plt.rcParams['figure.dpi'] = 300 #

plt.title("Message Latency")
plt.xlabel("Message id")
plt.ylabel("Message latency (ms)")
plt.show()
# plt.savefig('plottest.png', dpi=300) #
```
