```python
"""在csv文件中第一行添加索引字段"""
def write_raw_index(file):
    filename = file
    with open(filename, 'r+', encoding='utf-8') as f:
        content = f.read()
        f.seek(0, 0)
        #mid, text, source, uid
        text = "id" + "," + "message_latency"
        f.write(text + '\n' + content)
```

```python
filename = 'times_2000.csv'
filechangename = "time2000modify.csv"
with open(filename,"r+") as f:
    f_csv=csv.reader(f)
    with open(filechangename,"w+",newline='') as f_1:
        f_1_csv=csv.writer(f_1)
        f_1_csv.writerow(["id","message_latency"])
        for row in f_csv:
            intMessage = int(row[1])/1000000
            # print(intMessage)
            
            f_1_csv.writerow([row[0],str(intMessage)])
            
        f_1.close()
    f.close()
```

```python
fileplotname = "time2000modify.csv"
sample_data = pd.read_csv(fileplotname)
plt.plot(sample_data.id,sample_data.message_latency)


plt.rcParams['savefig.dpi'] = 300 #图片像素
plt.rcParams['figure.dpi'] = 300 #分辨率

plt.title("Message Latency")
plt.xlabel("Message id")
plt.ylabel("Message latency (ms)")
plt.show()
# plt.savefig('plottest.png', dpi=300) #指定分辨率保存
```
