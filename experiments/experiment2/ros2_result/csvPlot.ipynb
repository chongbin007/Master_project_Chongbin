```python
import  pandas as pd
import  matplotlib.pyplot as plt
import  csv
```

```python
x = [1,2,3]
y = [1,4,6]
z = [1,7,10]
```

```python
plt.plot(x,y)
plt.plot(x,z)


plt.title("Message Latency")
plt.xlabel("id")
plt.ylabel("latency")
plt.legend(["this y","this z"])
plt.show()
```

```python
sample_data1 = pd.read_csv('1.csv')
sample_data10 = pd.read_csv('10.csv')
sample_data1000 = pd.read_csv('1000.csv')
sample_data2000 = pd.read_csv('2000.csv')
```

```python
plt.plot(sample_data1.id,sample_data1.message_latency)
plt.plot(sample_data10.id,sample_data10.message_latency)
plt.plot(sample_data1000.id,sample_data1000.message_latency)
plt.plot(sample_data2000.id,sample_data2000.message_latency)
plt.legend(["1","10","1000","2000"])

plt.rcParams['savefig.dpi'] = 300 #图片像素
plt.rcParams['figure.dpi'] = 300 #分辨率

plt.title("Message Latency")
plt.xlabel("Message id")
plt.ylabel("Message latency (ms)")
# plt.show()
plt.savefig('plottest.png', dpi=300) #指定分辨率保存
```
