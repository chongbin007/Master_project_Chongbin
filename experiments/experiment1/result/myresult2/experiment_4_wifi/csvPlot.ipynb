```python
import  pandas as pd
import  matplotlib.pyplot as plt
import  csv
```

```python
x = [1,2,3]
y = [1,4,6]
z = [1,7,10]
```

```python
plt.plot(x,y)
plt.plot(x,z)


plt.title("Message Latency")
plt.xlabel("id")
plt.ylabel("latency")
plt.legend(["this y","this z"])
plt.show()
```

```python
sample_data = pd.read_csv('mean_times.csv')
```

```python
plt.plot(sample_data.id,sample_data.mean_time_ms_200)
plt.plot(sample_data.id,sample_data.mean_time_ms_400)

plt.plot(sample_data.id,sample_data.mean_time_ms_600)
plt.plot(sample_data.id,sample_data.mean_time_ms_800)

plt.plot(sample_data.id,sample_data.mean_time_ms_1800)
plt.plot(sample_data.id,sample_data.mean_time_ms_2000)
plt.legend(["1","10","600","800","1800","2000"])



plt.rcParams['savefig.dpi'] = 300 #图片像素
plt.rcParams['figure.dpi'] = 300 #分辨率

plt.title("Message Latency")
plt.xlabel("Message id")
plt.ylabel("Message latency (ms)")
plt.show()
# plt.savefig('plottest.png', dpi=300) #指定分辨率保存
```
